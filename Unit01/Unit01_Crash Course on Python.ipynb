### 01 - 奇偶數辨別（Odd or even）
#### 當使用者輸入一個的數字，它能夠辨別奇偶，並輸出檢驗結果給使用者。

> 題目：輸入一數字判斷奇數或偶數<br>
輸入(Input): 一個數字 <br>
輸出(Output):奇數或偶數

* 進階: 可以限制輸入一個介於一定範圍（例如 1 到 1000）的數字

````python
exceed = False
while(exceed == False):
    num = int(input("please enter a number "))
    
    if num > 0 and num <= 1000:
        if num % 2 == 0:
            print("偶數")
        else:
            print("奇數")
    else: 
        exceed = True
        print("不符合範圍")
````

### 02 - 填字遊戲（Mad libs game）
#### 使用者輸入任何字，它可以是名詞、形容詞、動詞、代名詞等。程式得到文字後，可自行排列，組成成一個模板故事。

> 題目：輸入一數字判斷奇數或偶數<br>

### 03 -  猜數字（Guess the number）
#### 請使用者猜一個介於範圍（例如 1 到 100）之間的數字，
#### 若使用者猜錯，就詢問他們想繼續玩還是退出；若使用者猜對，就顯示祝賀訊息，

![](data\guess.png)


````python
low = 1
high = 100

correct = False
while(correct == False):
    number = int(input("please enter an integer between 1 and 100 -> " ))
    if number >= low and number <= high:
        correct = True
    else: print("exceeds limitation!")

counter = 0
giveup = False
while(giveup == False):
    guess = int(input("please guess the hidden number -> "))
    counter += 1
    
    if guess < number:
        low = guess
    elif guess > number:
        high = guess
    
    if guess == number:
        print("you guessed it! Congratulations!")
        print(f"you have tried {counter} time(s)")
        break
    else: 
        want_to_quit = input("wrong answer! Do you want to try it again(Y/N)")
        if want_to_quit == 'N':
            print(f"you have tried {counter} time(s)")
            giveup = True
        elif want_to_quit == 'Y':
            giveup = False
            print(f"the range is {low} to {high}")
````


* 進階: 把file的學號抓出來，另存成一個新的csv檔

````python
import os
import csv

# 請使用者輸入或指定要查詢的路徑
yourPath = input()
# 列出指定路徑底下所有檔案(包含資料夾)
allFileList = os.listdir(yourPath)

# 逐一查詢檔案清單
for file in allFileList:
#   這邊也可以視情況，做檔案的操作(複製、讀取...等)
#   使用isdir檢查是否為目錄
#   使用join的方式把路徑與檔案名稱串起來(等同filePath+fileName)
    if os.path.isdir(os.path.join(yourPath,file)):
        print("I'm a directory: " + file)
    #   使用isfile判斷是否為檔案
    elif os.path.isfile(os.path.join(yourPath,file)):
        print("I'm a File: " + file)

serial = []
for file in allFileList:
    file = file.split('.')
    serial.append(file[0])
    
print(serial)

with open('student_serial.csv', 'w', newline = '') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=' ', quotechar='|', quoting=csv.QUOTE_MINIMAL)
    for s in serial:
        spamwriter.writerow(s)
````

